```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
covdat=pd.read_csv('C:\ML Algos\covidsymptomsdata.csv')
```

```python
covdat.head()
```

```python
covdat.info()
```

```python
covdat['Target'] = 0
covdat.head()
```

```python
conditions=[(covdat['Severity_None']==1),(covdat['Severity_Mild']==1),(covdat['Severity_Moderate']==1),(covdat['Severity_Severe']==1),]
values=[0,1,2,3]
covdat['Target']=np.select(conditions,values)
print(covdat)
```

```python
covdat.info()
```

```python
X = covdat.iloc[:20000,:18]
X.head()
```

```python
y = covdat.iloc[:20000,27]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
svmobj=SVC()
svmobj.fit(X_train,y_train)
y_pred=svmobj.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(f"Accuracy:-{accuracy_score(y_test,y_pred)}")
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=100)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
y_pred=knn.predict(X_test)
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
LR.score(X_test,y_test)
```
